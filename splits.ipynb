```python

import pandas as pd
import os
from pathlib import Path

base_dir = str(Path().absolute())
coarse_area = "/data/weather-stations/kelowna/kelowna_stations.csv"


metadata_file_path = base_dir + coarse_area
coarse_df = pd.read_csv(metadata_file_path)

coarse_df
```

```python
#calculate size of splits

splits = 12

min_latitude = 49.497
max_latitude = 51.0

w = max_latitude - min_latitude

min_longitude = -120.331 
max_longitude = -118.191

h = max_longitude - min_longitude

print(w)

print(h)

fine_width = w/splits

fine_height = h/splits

print(fine_width)
print(fine_height)

```

```python
# fine_df = split up 517 stations into manageable chunks
min_latitude = 49.497
max_latitude = 51.0

min_longitude = -120.331 
max_longitude = -118.191


fine_w = 0.125

fine_h = 0.178

splits = 12

for i in range(splits):
    # Create boolean masks for latitude and longitude conditions
    latitude_mask = (coarse_df['Latitude'] >= min_latitude)  & (coarse_df['Latitude'] <= min_latitude + fine_w) & (coarse_df['Latitude'] <= max_latitude)
    longitude_mask = (coarse_df['Longitude'] >= min_longitude) & (coarse_df['Longitude'] <= min_longitude + fine_h ) & (coarse_df['Longitude'] <= max_longitude)
    # Apply the masks to filter the DataFrame
    filtered_df = coarse_df[latitude_mask & longitude_mask]

    print(filtered_df)

    min_latitude += fine_w
    min_longitude += fine_h
    filtered_df.to_csv("fine_area"+ str(i) +".csv", index=False)



```
