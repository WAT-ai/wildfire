```python
import pandas as pd 
import os
```

```python
satellite_dir = "../data/processing/satellite-burn/"

file_path = satellite_dir + "satellite_burn_cleaned.csv"
df = pd.read_csv(file_path)
df
```

```python
def get_week_of_year(day):
    if pd.isna(day):
        return None
    return (day - 1) // 7 + 1

# Convert 'IGNITION_DATE' to datetime type for better handling
df['IGNITION_DATE'] = pd.to_datetime(df['IGNITION_DATE'])

# Group data by year and then by week
yearly_weekly_data_frames = {}
for (year, week), year_week_df in df.groupby([df['IGNITION_DATE'].dt.year, df['IGNITION_DATE'].dt.dayofyear.apply(get_week_of_year)]):
    if year not in yearly_weekly_data_frames:
        yearly_weekly_data_frames[year] = {}
    yearly_weekly_data_frames[year][week] = year_week_df

# Create a folder for each year and save CSV files within the folder
for year, weeks in yearly_weekly_data_frames.items():
    year_folder = f'{satellite_dir}{year}'
    os.makedirs(year_folder, exist_ok=True)
    
    for week, week_df in weeks.items():
        filename = f'week_{week}.csv'
        filepath = os.path.join(year_folder, filename)
        week_df.to_csv(filepath, index=False)
        print(f'{filename} exported to {year_folder} successfully.\n')

# Print the structure of the folders
for year in yearly_weekly_data_frames.keys():
    print(f'{year}_data:')
    for week in yearly_weekly_data_frames[year].keys():
        print(f'   week_{week}')
    print()
```
