```python
import pandas as pd
import os
```

```python
from pathlib import Path

base_dir = str(Path().absolute())
weather_stations_dir = "/data/weather-stations/"

metadata_file_path = base_dir + weather_stations_dir + "all_stations.csv"
meta_df = pd.read_csv(metadata_file_path)

meta_df
```

```python
# sample square of interest
top_left = (56, -125)
bottom_right = (49, -117.1105)

min_latitude = min(top_left[0], bottom_right[0])
max_latitude = max(top_left[0], bottom_right[0])
min_longitude = min(top_left[1], bottom_right[1])
max_longitude = max(top_left[1], bottom_right[1])

# Create boolean masks for latitude and longitude conditions
latitude_mask = (meta_df['Latitude'] >= min_latitude) & (meta_df['Latitude'] <= max_latitude)
longitude_mask = (meta_df['Longitude'] >= min_longitude) & (meta_df['Longitude'] <= max_longitude)

# Apply the masks to filter the DataFrame
filtered_df = meta_df[latitude_mask & longitude_mask]

filtered_df.to_csv("sample_filtered_stations.csv", index=False)
```
